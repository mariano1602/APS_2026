# Cuantizacion de una senoidal con 3 bits
## Mariano Gonzalez

```python
import numpy as np
def mi_funcion_sen( vmax, dc, f0, ph, N, fs):
    tt=np.arange(0,N,1)*(1/fs)
    xx=dc + vmax*np.sin(2*np.pi*f0*tt + ph)
    return tt,xx 

import matplotlib.pyplot as plt

tt, xx = mi_funcion_sen( vmax = 1, dc = 0, f0 = 1, ph=0,N=100, fs = 100)

plt.figure()
plt.grid()
plt.plot(tt, xx)
plt.show()
```

# Cuantizacion 

```python
b=3 # b-bits
vfs=3 #volts
qq=vfs/2**b
xxq=np.round(xx/qq) 

plt.figure()
plt.grid()
plt.plot(tt, xxq)
plt.show()
```

El programa consiste en desarrollar una función que genere señales senoidales y que permita parametrizar:

-la amplitud máxima de la senoidal (volts)

-su valor medio (volts)

-la frecuencia (Hz)

-la fase (radianes)

-la cantidad de muestras digitalizada por el ADC (# muestras)

-la frecuencia de muestreo del ADC.

Declaramos un funcion Seno y graficamos una senoidal de 1Hz:


```python
import numpy as np

def mi_funcion_sen( vmax, dc, f0, ph, N, fs):
    tt=np.arange(0,N,1)*(1/fs)
    xx=dc + vmax*np.sin(2*np.pi*f0*tt + ph)
    return tt,xx 
import matplotlib.pyplot as plt

tt, xx = mi_funcion_sen( vmax = 1, dc = 0, f0 = 1, ph=0,N=100, fs = 100)
plt.figure()
plt.grid()
plt.plot(tt, xx)
plt.show()
```

 ## Teorema de Nyquist

El **Teorema de Nyquist-Shannon** establece que:

> *Para poder reconstruir correctamente una señal analógica a partir de sus muestras digitales, la frecuencia de muestreo debe ser al menos el doble de la frecuencia máxima presente en la señal.*

Matemáticamente:

\begin{equation}
f_s \geq 2 \cdot f_{\text{max}}
\end{equation}

donde:

-$f_s$ = frecuencia de muestreo  
-$f_{\text{max}}$ = frecuencia máxima de la señal  

Si no se cumple esta condición, aparece el fenómeno de **aliasing**, en el cual las frecuencias altas se confunden con otras más bajas en la señal muestreada.
Por lo tanto, debemos aumentar Fs a 2000Hz, que es el doble de 1000, justo en el punto donde podemos ver el fenomeno de **aliasing**

# Ruido

Como la señal tiene propiedad aditiva, cualquier perturbacion indeseada puede sumarse a la onda original, esta perturbacion puede ser el ruido

```python
N=100
ruido=np.random.normal(0,0.1,N)
ttr, xxr = mi_funcion_sen( vmax = 1, dc = 0, f0 = 1, ph=0,N=100, fs = 100)
yyr=xxr + ruido
plt.figure()
plt.grid()
plt.plot(ttr, yyr)
plt.show()
```

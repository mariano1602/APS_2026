#### b) Graficar Modulo y Fase

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft

N=8
n=np.arange(N)
dc=4
vmax=3
xx= dc + vmax*np.sin((np.pi/2)*n)

# %% FFTs
XX=fft(xx)
XXabs=np.abs(XX) 
XXang=np.angle(XX)

##graficamos modulo
plt.figure(1)
plt.clf()
plt.grid()
plt.plot(n,XXabs, "o")
plt.title("Modulo")
plt.ylabel('Amplitud del modulo')
plt.xlabel('n')
# plt.legend()
plt.show()

##graficamos fase
plt.figure(2)
plt.clf()
plt.grid()
plt.title("Fase")
plt.ylabel('Fase  [rad]')
plt.xlabel('n')
plt.plot(n,XXang, "o")
plt.show()
```

## 3) validacion de respuesta en frecuencia de todos los sistemas con Numpy.

```python
from scipy.signal import freqz
import numpy as np
import matplotlib.pyplot as plt
```

### Validacion punto a)

```python
b = [1,1,1,1]   #a)
a = [1]       # denominador

w, H = freqz(b, a, worN=1024)
plt.clf()
plt.figure(1)
plt.plot(w, np.abs(H),  label = 'Amplitud')
plt.plot(w, np.angle(H), label = 'Fase')
plt.xlabel('Frecuencia [rad/muestra]')
plt.ylabel('∠H(w) y |H(w)|  [rad]')
plt.grid(True)
plt.legend()
plt.show()

```

#### Podemos ver que en modulo representa la curva predicha analiticamente, se ve claramente una especie de lobulo principal y un secundario. Filtro pasa bajo. En fase los valores en omega=0 y omega=pi condice con lo lo resuelto analiticamente, en omega=pi/2 ocurre un desfasaje de 90°, vuelve al inicio.

Luego vemos los salto de fase en los ceros de modulo.
